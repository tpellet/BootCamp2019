### Exercise 1

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('titanic.csv')
data
```

```python
data[data['Survived'] == 1].plot(kind='hist', y ='Age', title="Age distribution of \
surviving Titanic passengers", stacked = True)
data[data['Survived'] == 0].plot(kind='hist', y ='Age', title="Age distribution of \
dead Titanic passengers", stacked = True, color='orange')
plt.xlabel('Age (years)')
plt.title
```

```python
pd.plotting.scatter_matrix(data, figsize=(12,12))
```

```python
data.drop(['Body', 'Fare', 'Age'], axis=1).plot(kind="box", vert=False, title='Box plot for variables related to the Titanic Dataset')
```
