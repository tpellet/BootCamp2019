```python
import pandas as pd
from matplotlib import pyplot as plt
import pydataset as data

```

```python
iris = data.data('iris')
iris.info()
species = iris.groupby('Species')
list(species)

species.describe()
```

The Virginica is much bigger than the other two

```python
species.boxplot(grid=False)
plt.tight_layout()
```

```python
iris.boxplot(['Sepal.Length','Petal.Length'], by="Species", grid=False)
```

To distinguish between the Setosa and the other two, one could compute the ratio of Sepal to Petal length. To distinguish between Vericolor and Virginica, one could compare average sizes across all 4 measures. 

```python
poisons = data.data('poisons')
data.data('poisons', show_doc=True)
```

```python
poisons.boxplot(['time'], by="poison", grid=False)
```

```python
poisons.boxplot(['time'], by="treat", grid=False)
```

The poison 3 seems to be the most effective while treatment B tends to increase the likelihood of surviving. 

```python
poison_treat = poisons.groupby(['poison', 'treat'])
poison_treat.boxplot(['time'], grid=False, figsize= (20,20), vert=False)
```

```python
diamonds = data.data('diamonds')
data.data('diamonds', show_doc=True)
diamonds
```

```python
color = diamonds.groupby(['color'])
diamonds.boxplot(['price'], by='color', grid=False, vert= False, figsize= (20,20))
```

```python
diamonds.boxplot(['price'], by='cut', grid=False, vert= False, figsize= (20,20))
```

The color and cut affect the tail of the distribution most. Good color and good cut will increase the number of expensive diamonds, without necessarily increasing their average value. 

```python
diamonds[diamonds['color'] == 'H'].drop(['price'], axis=1).boxplot(by='cut', grid=False, vert= False, figsize= (20,20))
```

Fair diamonds do not have the same shape as the others. Their depth and table can be slightly larger.

```python
titanic = pd.read_csv('titanic.csv')
titanic.pivot_table(values = "Survived", columns="Embarked", aggfunc='mean')
```

```python
titanic.pivot_table(values = "Survived", index='Sex', columns="Embarked", aggfunc='mean')
```

```python
titanic.pivot_table(values = "Survived", index=['Embarked', 'Pclass'], columns="Sex", aggfunc='count')
```

High correlation between location and passenger class suggest that the class aspect is driving the correlationb between location and mortality. 

```python
age = pd.cut(titanic['Age'], [0, 12, 18, 80])
titanic.pivot_table(values = "Age", index=['Embarked', 'Pclass'], columns=['Sex'], aggfunc='mean')
```
